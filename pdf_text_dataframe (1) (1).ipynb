```python
def move_files(string_to_match, source_folder, dest_folder, dest_folder1):
    for filename in os.listdir(source_folder):
        proof = 2
        if filename[-3:] == 'txt':
            text = open(source_folder+filename, 'r')
            #print(type(text.readlines()))
            txt2 = ' '.join(text.readlines())
            text.close()
            for i in string_to_match:
                if i in txt2:
                    shutil.move(os.path.join(source_folder, filename), dest_folder)
                    proof = 0
                    break
            if proof > 0:
                shutil.move(os.path.join(source_folder, filename), dest_folder1)    
```

```python
import os
import shutil
string_to_match = ['with one exception','with the exception','have not been met','will not be met to the extent', 'will not be met','following exception','deviations','certain derogations']
source_folder=r'E:/New folder/'
dest_folder = r'E:/New folder/exception/'
dest_folder1 = r'E:/New folder/complied/'
move_files(string_to_match, source_folder, dest_folder, dest_folder1)
```
